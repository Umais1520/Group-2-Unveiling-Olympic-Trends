# Importing relative libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
import warnings
warnings.filterwarnings ('ignore')
```

# Importing Data

```python
df1= pd.read_csv('athlete_events.csv') #Rather than importing the files from an external data base, the files were uploaded onto jupyter notebooks.
df2= pd.read_csv('noc_regions.csv')   # They were then imported into python by the followi g formula
```

```python
Data= pd.merge(df1,df2, on= 'NOC', how='inner') #the two files have been merged into one single file called data.
```

# Data Cleaning

```python
#Data Cleaning
Data.rename(columns={'region': 'Region', 'notes' : 'Notes'}, inplace=True) #region and notes were the only names in lowercase letter. we have converted them to uppercase for visual clarity
```

```python
Data.head()
```

```python
new_order = ['ID','Name', 'Age', 'Height', 'Weight', 'Sex', 'NOC', 'Region', 'Year', 'Season', 'City', 'Sport', 'Event', 'Medal','Notes','Games','Team']
Data= Data[new_order]
Data.head()

#columns are rearrnaged to make better sense. NOC and region are placed next to each other beacuse not everyone knows short form for countries.
```

```python
Columns_to_drop= ['Games', 'Team']
Data= Data.drop(columns=Columns_to_drop)
Data.head()
#The columns of games and team have been dropped. info of game is given in year and season column. Team and region also give the same info
```

```python
Data.dtypes
#identified different types of data as it will help in further analyses
```

```python
missing_values_per_column = Data.isna().sum()
print(missing_values_per_column)
#identified the number of missing values. It will help in analysing the data after removing the missing values
```

```python
Data['Age']= Data['Age'].fillna(Data['Age'].mean())
Data
#calculating age through mean helpls us fill missing values without introducing extreme values. moreover it avoids removing rows for the sake of omitting missing data.
```

```python
missing_values_per_column = Data.isna().sum()
print(missing_values_per_column)
#we can now see that the age column has no missing values
```

```python
Columns_to_fill= ['Height', 'Weight','Notes']
Data[Columns_to_fill]= Data[Columns_to_fill].fillna(0)
Data.head()
#assigned 0 to the missing values in the column of height, weight, and notes.
```

```python
Columns_to_fill= ['Medal']
Data[Columns_to_fill]= Data[Columns_to_fill].fillna("No Medal Won")
Data.head()
#Filled the missing values in Medals column with 'No medals won'.
```

```python
Data
```

```python
Data['Age'] = Data['Age'].astype(int)
Data['Height'] = Data['Height'].astype(int)
Data['Weight'] = Data['Weight'].astype(int)
Data.head()
#converted the data in age, height and weight column as whole numbers. seems visually appealing.
```

# Data Wrangling

```python
#Data Wrangling
bins = [0,18,25,35, float('inf')] 
labels = ['0-18' , '19-25', '26-35' , '36+']

Data['Age Group'] = pd.cut(Data['Age'], bins=bins, labels=labels, right=True)
Data
#we have created a new column neamed 'age group'.
```

```python
Data['Surname']= Data['Name'].apply(lambda x: ' '.join(x.split()[1:])
                                    if len(x.split()) > 1 
                                    else x)
Data.head()
#created a new column for surname where surname is dispalyed for those who have surname only and middle name along with surname for clarity if someone has the same surname.
```

```python
Data['Year'] = pd.to_numeric(Data['Year'], errors= 'coerce')
Data['Century'] = Data['Year'].apply(lambda x: (x // 100+1) 
                                     if pd.notna(x)
                                     else None)
Data.head()
#We first converted the year column into numerical data and then applied the function to extract the century.
```

```python
medal_counts = Data.groupby('Region')['Medal'].count().reset_index()
most_medals_region = medal_counts.loc[medal_counts['Medal'].idxmax(), 'Region']
Data['Most Medals Won']= most_medals_region
Data
#created a new column called most medals won and calculated which region has the most number of medals
```

```python
sports_century_counts = Data.groupby(['Century', 'Age Group'])['Medal'].count().unstack()
sports_century_counts.plot(kind='line', figsize=(12,6), title= ' Age Group Performance Across Centuries')
plt.xlabel('Century', color='blue')
plt.ylabel('Number of Medals', color= 'red')
plt.legend(title='Age Group')
plt.show()
           
```

This code visualizes the relationship between age and medal counts using a scatter plot. It also highlights the age group with the highest number of medals by adding an annotation to the chart. The scatter plot provides a clear visual representation of how medal counts vary with age. The annotation highlights the specific age group that won the highest number of medals (Age 25 with 200 medals in this example).

```python
gender_medal_counts = Data.groupby(['Sex', 'Medal']).size().unstack()
gender_medal_counts.plot(kind='bar', stacked=True, title='Medals by Gender', figsize=(10, 6))
plt.xlabel('Gender')
plt.ylabel('Number of Medals')
plt.legend(title='Medal Type')
plt.show()
```

This code analyzes and visualizes the distribution of medals by gender using a stacked bar chart. The total height of each bar represents the total number of medals won by each gender. It is identifiable which gender has won more medals overall. The stacked sections within each bar show the proportion of Gold, Silver, and Bronze medals for each gender. This helps identify whether one gender excels in winning a specific type of medal (e.g., are males winning more Gold medals than females?). The visualization can reveal whether medal distributions are balanced between genders or if there are noticeable disparities.

```python
sport_medal_counts = Data.groupby(['Sport','Medal']).size().unstack()
sport_medal_counts.plot(kind='bar', stacked=True, title='Medals by Sport', figsize=(18, 6))
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
plt.legend(title='Medal Type')
plt.show()
```

```python
region_medal_counts = Data.groupby(['Region', 'Medal']).size().unstack()
region_medal_counts.plot(kind='line', stacked=True, title='Medals by region', figsize=(18, 6))
plt.xlabel('Region')
plt.ylabel('Number of Medals')
plt.legend(title='Medal Type')
plt.show()
```

```python
participation = Data.groupby('Year')['ID'].nunique().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='ID', data=participation, marker='o')
plt.title('Number of Athletes Participating Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Athletes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
countries = Data['NOC'].unique()
print("Available countries:")
print(", ".join(countries))
country = input("\nEnter the NOC code of the country you want to analyze: ").strip().upper()
if country not in countries:
    print(f"\nError: '{country}' is not a valid NOC code. Please try again.")
else:   
    country_medals = Data[Data['NOC'] == country].groupby('Year')['Medal'].count().reset_index()
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='Year', y='Medal', data=country_medals, marker='o',color='royalblue')
    plt.title(f'Medal Trends for {country} Over Time',color='darkblue')
    plt.xlabel('Year',color='red')
    plt.ylabel('Number of Medals',color='red')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
sport_participation = Data.groupby('Sport').size().reset_index(name='Participation')
most_participation_sport = sport_participation.sort_values(by='Participation', ascending=False)
most_participation_sport.head()
```

```python
plt.figure(figsize=(15, 8))  # Chart 
plt.plot(most_participation_sport['Sport'], most_participation_sport['Participation'], 
         marker='o', linestyle='-', linewidth=2, color='blue')  # Line Chart

# Chart Details
plt.title('Participation by Sport (Line Plot)', fontsize=16)  # 
plt.xlabel('Sport', fontsize=12)  # X-axis Label
plt.ylabel('Number of Participants', fontsize=12)  # Y-axis Label
plt.xticks(rotation=90, fontsize=10)  # X-axis Label  
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Y-axis Gridlines
plt.tight_layout()
plt.show()
```
